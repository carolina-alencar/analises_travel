## Importando arquivo cvs

```python
import pandas as pd
```

```python
dados = pd.read_csv('Travel_details_dataset.csv')
```

```python
dados
```

## Alterando o nome das colunas 

```python
new_names = ['id','destino','data_inicio','data_fim','duração_dias','nome','idade','sexo','nacionalidade','tipo_acomodacao','custo_acomodacao','tipo_transporte','custo_transporte']

def renomear(df:pd.DataFrame, lista_nomes: list) -> None:
    for velho_nome, novo_nome in zip(list(df.columns),lista_nomes):
        print(f'{velho_nome}:{novo_nome}')
        df.rename(columns={velho_nome: novo_nome}, inplace = True)

renomear(dados,new_names)
```

```python
dados.head(11)
```

## Covertendo as colunas de custo para float

```python
dados['custo_acomodacao'] = dados['custo_acomodacao'].str.replace('[^\d]', '', regex=True).astype(float)
```

```python
dados['custo_transporte'] = dados['custo_transporte'].str.replace('[^\d]', '', regex=True).astype(float)
```

```python
dados.info()
```

### Convertendo coluna de data para tipo date

```python
dados['data_inicio'] = pd.to_datetime(dados['data_inicio'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
dados['data_fim'] = pd.to_datetime(dados['data_fim'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
```

```python
dados['data_inicio']
```

### Tratando dados nulos 


```python
dados = dados.fillna(0)
```

```python
dados.isnull().sum()
```

```python
remover_nulos = dados.query('destino == 0 or data_inicio == 0 ').index
remover_nulos
```

```python
dados.drop(remover_nulos, axis=0, inplace=True)
```

### Analisando os dados

### Agrupando os dados por ano

```python
dados['data_inicio'] = pd.to_datetime(dados['data_inicio'])  # Converte a coluna de datas para o tipo datetime
dados['ano_inicio'] = dados['data_inicio'].dt.year  # Cria uma nova coluna 'ano' com o ano da data
```

```python
# contagem de Destinos pelos anos
dados_ano = dados.groupby(['ano_inicio', 'destino'])['destino'].count().reset_index(name='contagem')
dados_ano
```

```python
# Quantidades de destinos por ano
dados_ano_cont = dados_ano.groupby('ano_inicio')['contagem'].sum()
dados_ano_cont
```

### Média de duração das viajens por destino

```python
media_duracao = dados.groupby('destino')[['duração_dias']].mean().sort_values(by='duração_dias',ascending=False)
media_duracao['duração_dias'] = media_duracao['duração_dias'].round(1)
media_duracao.head()
```

### Percentual do genero Feminino e Masculino

```python
# Alterando os nomes dos generos:
dados['sexo'] = dados['sexo'].replace({'Female': 'F', 'Male': 'M'})
```

```python
percentual_genero = dados['sexo'].value_counts(normalize=True).to_frame().sort_values('sexo').round(2)
percentual_genero 
```

### Média de idade por sexo

```python
media_idade_sexo = dados.groupby('sexo')[['idade']].mean().sort_values('idade').round(1)
media_idade_sexo
```

### Destinos mais frequentados por nacionalidade

```python
# Agrupar os dados por nacionalidade e destino e contar a frequência
destino_por_nacionalidade = dados.groupby(['nacionalidade', 'destino']).size().reset_index(name='contagem')

# Encontrar o destino mais frequente para cada nacionalidade
destino_mais_frequente = destino_por_nacionalidade.groupby('nacionalidade')[['destino', 'contagem']].max().sort_values(by='contagem', ascending=False)

destino_mais_frequente.head(10)
```

### Média de idade por sexo e destino

```python
pivot_table = dados.pivot_table(values='idade', index='sexo', columns='destino', aggfunc='mean').round(1)
pivot_table = pivot_table.fillna(0)
pivot_table.iloc[:, :10]
```

### Tipos de hospedagem mais utilizadas por destino

```python
hospedagem_destino = dados.groupby(['destino', 'tipo_acomodacao']).size().reset_index(name='contagem')
hospedagem_destino = hospedagem_destino.sort_values(by='contagem', ascending=False)
hospedagem_destino.head(10)
```

### Os 10 destinos com maior média de custo de hospedagem.

```python
media_hospedagem_destino = dados.groupby('destino')[['custo_acomodacao']].mean().sort_values('custo_acomodacao',ascending=False).head(10)
media_hospedagem_destino.plot(kind='barh', figsize=(9, 8), color = 'pink')
```

### Transportes mais utilizados por destino

```python
transporte_destino = dados.groupby(['destino', 'tipo_transporte']).size().reset_index(name='contagem')
transporte_destino = transporte_destino.sort_values(by='contagem', ascending=False)
transporte_destino.head(10)
```

### Os 10 destinos com maior média de custo com transporte

```python


media_transporte_destino = dados.groupby('destino')[['custo_transporte']].mean().sort_values('custo_transporte',ascending=False).head(10)
media_transporte_destino.plot(kind='barh', figsize=(9, 8), color = 'pink')
```

### Coluna categórica


```python
# Descrição do viajante
dados['descricao'] = 'viajante: ' + dados['nome'] + ', nacionalidade: ' + dados['nacionalidade'] + ', viajou para: ' + \
                                        dados['destino'] + ', por ' + \
                                          dados['duração_dias'].astype(str) + ' dias, ' + 'hospedagem em: ' + \
                                            dados['tipo_acomodacao'] + ', meio de transporte: ' +  dados['tipo_transporte'].astype(str)
dados.Descricao.head()
```
